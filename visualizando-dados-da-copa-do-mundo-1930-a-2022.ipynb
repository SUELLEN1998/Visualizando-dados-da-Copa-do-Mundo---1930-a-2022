Utilizando o dataset Kaggle - https://www.kaggle.com/datasets/martj42/international-football-results-from-1872-to-2017?select=results.csv

```python
import pandas as pd
```

```python
results = pd.read_csv('/content/drive/MyDrive/Arquivos de DataSets/results.csv.zip', parse_dates=['date'])
```

```python
results.info()
```

```python
results_fifa = results[results['tournament'] == 'FIFA World Cup']
```

```python
results_fifa
```

```python
results_fifa['year'] = results_fifa['date'].dt.year
```

```python
results_fifa
```

```python
home_scores = results_fifa[['home_team', 'home_score', 'year']].rename({'home_team':'team', 'home_score':'score'}, axis='columns')
away_scores = results_fifa[['away_team', 'away_score', 'year']].rename({'away_team':'team', 'away_score':'score'}, axis='columns')
```

```python
home_scores
```

```python
away_scores
```

```python
scores_fifa = pd.concat([home_scores, away_scores]).reset_index(drop=True).fillna(0)
```

```python
scores_fifa
```

```python
scores_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0)
```

```python
score_fifa_year = scores_fifa.groupby(['team', 'year']).score.sum().unstack().fillna(0).cumsum(axis=1)
```

```python
flags = pd.read_csv('/content/drive/MyDrive/Arquivos de DataSets/countries-fifa-flags.csv')
```

```python
flags.info
```

```python
score_fifa_year
```

```python
score_fifa_year.merge(flags, left_on='team', right_on='country', how='left')
```
